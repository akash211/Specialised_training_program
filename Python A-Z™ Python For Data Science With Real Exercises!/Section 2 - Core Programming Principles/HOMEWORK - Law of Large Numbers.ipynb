```python
#Solution 1:
    
import numpy as np
from numpy.random import randn

N = 100000
counter = 0

for i in randn(N):
    if (i > -1 and i < 1):
        counter = counter + 1

print(counter / N)
```

```python
#Solution 2:
    
import numpy as np
from numpy.random import randn

N = 100000
counter = 0

for i in range(N):
    temp = randn()
    if (temp > -1 and temp < 1):
        counter = counter + 1

print(counter / N)
```

```python
#Solution 3 (taking user input for N and by creating function):

from numpy.random import randn

N = int(input('Enter value of N: - '))

def normal(N):
    counter = 0   #can not keep outside the function
    
    for i in randn(N):
        if (i <= 1 and i >= -1):
            counter += 1

    print(counter/N)

normal(N)
```
